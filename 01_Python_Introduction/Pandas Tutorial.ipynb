# Pandas Series and Pandas Dataframe
* Before we start dealing with some of Pandas’ tools, we need mention the two data structures Pandas uses to store data, the Pandas **Series** and the Pandas **Dataframe**.

* Think of Pandas Series as an 1 column Excel spreadsheet, with an additional index column, or even better, if you are familiar with Numpy think of an one dimensional array.


### Intorduction to Series

```python
wind_speed = pd.Series([0.5, 1.2, 2.1, 0.3, 0.9, 1.4, 2.8])
print(wind_speed)
```

* We use the **pd.Series()** command, and provide a list

```python
# check datatype of wind_speed
type(wind_speed)
```

* We also can create a Series from an 1-dim Numpy array. 

```python
import numpy as np


array = np.array([0.5, 1.2, 2.1, 1.1, 1.6, 1.8, 1.4, 
                        0.6, 0.4, 0.3, 0.9, 2.5, 2.1, 2.8])

wind_speed_b = pd.Series(array)
print(wind_speed_b)
print(type(wind_speed_b))
```

### Introduction to DataFrame

* what if we have multiple Series, what if we also have for example precipitation measurements, well then, Pandas stores our data in a Dataframe. 

* A Dataframe is just a container for multiple Series.


*  Think of it as an Excel spreadsheet with multiple columns, and one index column, or as an N-dimensional Numpy array.


###### Example

```python
array_ws = np.array([0.5, 1.2, 2.1, 1.1, 1.6, 1.8, 1.4, 
                        0.6, 0.4, 0.3, 0.9, 2.5, 2.1, 2.8])
array_precip = np.array([0., 0., 5.8, 4.3, 2., 1.1, 0., 
                        0.7, 0.9, 1.3, 1.9, 6.5, 8.2, 1.8]) 
array_temp = np.array([11., 12.1, 12.2, 12.7, 10.8, 9.7, 13.2,
                      12.8, 12.6, 14.1, 14.8, 15.3, 14.4, 16.2])

weather_df = pd.DataFrame({"wind_speed": array_ws,
                          "precipitation": array_precip,
                          "temperature": array_temp})

weather_df
```

* To create a DataFrame we use the pd.DataFrame() command, and provide a dictionary where the keys represent our column names, and the values are the before created Numpy arrays (our data)

# Missing values

* we are likely to encounte issing values. In order to maintain data accuracy and consistency, it is of crucial importance to detect and handle missing values appropriately.

```python
melb_df.isna().sum()    
```

# Handling missing values

* The fillna function can be used to replace the missing values. There are different approaches to handle them. The optimal one depends on the data at hand.

* We can fill the missing values with the mean, median, or the mode of the values in a column. Another option is to use a constant to replace the missing values.


#### drop na
* In some cases, we can choose to drop the rows or columns that have missing values. Thedropna function is used for such tasks. It also allows for dropping rows with missing values based on a condition

```python
melb_df.dropna(axis=0, how="any", inplace=True) 
melb_df["Price"].isna().sum()
```

* The **axis** parameter selects rows or columns

```python
melb_df.shape
```

#### After dropping missing values out dataset reduce to 6, 196 data.

## astype() Function
* The astype function can be used for changing the data type of a column. For instance, I may choose to represent the prices as integers rather than floats.

```python
melb_df.Price = melb_df.Price.astype("int")
melb_df.Price.dtypes
```

# Mathematical operations on column
* A common operation in data analysis is to derive columns based on the existing ones. We can either update the values in a column or create a new one off of an existing one.

```python
melb_df["PriceMillion"] = melb_df["Price"] / 1000000
melb_df.head()
```

# Filtering Observation

* Filtering observations (i.e. rows) is a fundamental part in data analysis. We try to obtain a better understanding of the data by checking it based on some conditions.

```python
melb_df[melb_df.PriceMillion > 1]
```

# Filtering on textual data
* Pandas provides several functions and methods to work with textual data. We can use them through the str accessor.

```python
suburb_D = melb_df[melb_df.Suburb.str.startswith("D")]
suburb_D.head()
```

# Groupby

* The groupby function is frequently used in data analysis. We can compare the groups or distinct values in a column based on some other values.

* For instance, we can calculate the average house price in each region group. We select the columns of interest and specify the one used for grouping. Then, an aggregation function is applied to each group.



```python
melb_df[["Suburb", "Price"]].groupby("Suburb").mean()
```

# Rank
* The rank function is used for assigning a rank to the observations (i.e. rows) based on the values in a column. For instance, we can rank the houses based on the price.

```python
melb_df["Rank"] = melb_df["Price"].rank(method="first",  
                              ascending=False).astype("int")
melb_df[["Price", "Rank"]].head()
```


# sort_values

* The sort_values function sorts the rows based on the values in a given column. Let’s use it to sort the houses based on price.

```python
melb_df.sort_values(by="Price", ascending=False)[["Price", "Rank"]].head()
```

Removing duplicate entries can be easily done with help of the pandas function.  # Concatenating DataFrame
 
 * In order to concat a dataframe, we use .concat() function this function concat a dataframe and returns a new dataframe.

```python
# Define a dictionary containing employee data 
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']} 
   
# Define a dictionary containing employee data 
data2 = {'Name':['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'], 
        'Age':[17, 14, 12, 52], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
# Convert the dictionary into DataFrame  
df = pd.DataFrame(data1,index=[0, 1, 2, 3])
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data2, index=[4, 5, 6, 7])


# using a .concat() method
frames = [df, df1]
 
res1 = pd.concat(frames)
res1
```

# Merging a dataframe 
* Merging a dataframe with one unique key combination

```python

 
# Define a dictionary containing employee data 
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 
   
# Define a dictionary containing employee data 
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
 
# Convert the dictionary into DataFrame  
df = pd.DataFrame(data1)
 
# Convert the dictionary into DataFrame  
df1 = pd.DataFrame(data2) 
```
