<h1 class="h1">Fast API Tutorial: A Framework to Create APIs</h1>
<p>In this tutorial, we will learn about the FastAPI framework and how we can create REST APIs. We will learn why FastAPI is so popular and how it differs from the other available frameworks. We will also discuss how to create web APIs that implement best practices by default quickly. By the end of this tutorial, we will be able to start creating production-ready web APIs and be able to go deeper for specific cases. We assume that you are familiar with HTTP and how it works, JSON, and Python basics. So let's move to the introduction of the FastAPI.</p>
<h2 class="h2">What is FastAPI?</h2>
<p>FastAPI is a popular, modern, and high-performance web-framework for creating web services or APIs. It is a Python based framework comes with the standard type hints. REST is a set of the rules that outlines the best practices for sharing data between clients and servers. As its name suggest, it is fast due to its out-of-the-box support of the async feature of the Python 3.6+.</p>
<p>This framework was released in 2018 and it was created by the Sebastián Ramírez. An interesting reason behind creating this project, he was not happy with the current available framework like Flask and DRF so he created own framework using tools such as Starlette and Pydantic.</p>
<p>Now, FastAPI is quite popular and demanding in industry even big tech giants like Uber, Netflix, and Microsoft are using FastAPI to build their apps.</p>
<h2 class="h2">FastAPI Features</h2>
<p>FastAPI provides many features over the currently available framework. These features are given below.</p>
<ul class="points">
<li><strong>High-performance -</strong> FastAPI is quite fast. It is considered one of the fastest frameworks currently present. It is par with NodeJS and Go, and also Starlette and Pedantic.</li>
<li><strong>Fast to code -</strong> It allows for significant increases in development speed.</li>
<li><strong>Robust -</strong> It has a well-written document, and we can create the production-ready APIs using our own by reading its documentation.</li>
<li><strong>Intuitive -</strong> It is designed as everyone can understand easily, is easy to use and offers excellent editor support.</li>
<li><strong>Fewer Bugs -</strong> It reduces around 40% of induced bugs. </li>
<li><strong>Plugins -</strong> It provides the facility to create the plugins using dependency injection.</li>
<li><strong>Type Hints -</strong> We can use type hinting for data validation and conversion.</li>
<li><strong>Standard Based -</strong> It is based on the open standards for APIs, OpenAI and JSON Schema. </li>
<li><strong>Straightforward -</strong> It is straightforward, so we can spend less time reading documentation. </li>
</ul>
